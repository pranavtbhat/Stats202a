```python
%matplotlib inline
import StagewiseRegression as sw
import numpy as np
import matplotlib.pyplot as plt
```

# Build Random Dataset

```python
n = 50
p = 200
s = 10

X = np.random.randn(n,p)

beta_true = np.zeros(p)
beta_true[0:s] = range(1, s + 1)

Y = X.dot(beta_true) + np.random.randn(n)
```

# Calculate **beta_all**

```python
beta_all = sw.swRegression(X, Y)
```

# Plot Stagewise regression solution path

```python
plot_x = np.transpose(np.ones((1,p)).dot(abs(beta_all)))
plot_y = np.transpose(beta_all)

plt.title("Trace of Stagewise regression solution")
plt.plot(plot_x, plot_y)

plt.ylabel("All Beta Values")
plt.xlabel("Iterations")
```

Stagewise regression is a form of monotone lasso regression, in that it doesn't allow a sudden change in curve direction. A beta value can only be increased monotonically. The magnitues of beta increase with each iteration, or remain stagnant. Stagewise regression doesn't penalize beta values. The stagewise regression therefore tends to take fewer iterations than the lasso regression and shows a lower error.
