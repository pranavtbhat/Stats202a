# Prepare dataset

```r
n = 50
p = 200
s = 10

X = matrix(rnorm(n * p), nrow=n)

beta_true = matrix(rep(0, p), nrow=p)
beta_true[1 : s] = 1 : s

Y = X %*% beta_true + rnorm(n)
```

```r
beta_all = swRegression(X, Y)
```

```r
num_iter = t(matrix(rep(1, p), nrow = 1) %*% abs(beta_all))
Beta_all = t(beta_all)

matplot(num_iter, Beta_all, type = 'l')
```
