# Lasso Regression Solution Path in Python

Lasso (least absolute shrinkage and selection operator) regression, is better than ridge regression, especially for the one dimensional case. Lasso regression both selects variables and regularizes the data, to improve prediction accuracy. 

```python
import Lasso
```

## Randomly generated data

```python
##############################
## Plotting
##############################
n = 50
p = 200
s = 10

X = np.random.randn(n,p)
lambda_all = list(range(1000, 0, -10))
beta_true = np.zeros(p)
beta_true[0:s] = list(range(1, s + 1))

Y = np.dot(X, beta_true) + np.random.randn(n)
```

## Solve for $\beta$ starting with a large $\lambda$ value 

```python
beta_all = Lasso.myLasso(X, Y, lambda_all)
```

## Plot of $\beta$ values over iterations

```python
import matplotlib.pyplot as plt

M = np.transpose(np.dot(np.ones((1,p)), abs(beta_all)))
betas = np.transpose(beta_all)

plt.title("Trace of Lassso Solution")
plt.plot(M, betas)
plt.ylabel("All beta values")
plt.xlabel("N")
```

As can be seen from the plot obtained, Lasso regression prefers a sparse $\beta$. Most beta values remain zero throughout. Iterations of the Coordinate descent cause new variables to be selected, and in some cases, variables to be dropped. However, beyond a certain number of iterations, forward selection doesn't work as well. As shown in the graph, most variables develop small but non-zero of beta, indicating that most variables have been selected.
